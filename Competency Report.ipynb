```python
class NodeLL:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, item):
        if not self.head:
            self.head = NodeLL(item)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = NodeLL(item)
    def insert(self, index, value):
        if not self.head:
            self.head = NodeLL(value)
            
        elif index <= 0:
            new_node = NodeLL(value)
            new_node.next = self.head
            self.head = new_node
            
        else:
            prev = None
            current = self.head
            counter = 0
            while current.next and counter != index:
                prev = current
                current = current.next
                counter += 1
            if counter != index:
                current.next = NodeLL(value)
            else:
                new_node = NodeLL(value)
                prev.next = new_node
                new_node.next = current        
    def delete_node(self, key):
        curr = self.head
        prev = None
        while curr and curr.data != key:
            prev = curr
            curr = curr.next
        if prev is None:
            self.head = curr.next
        elif curr:
            prev.next = curr.next
            curr.next = None

    def is_empty(self):
        return self.head == None
    
    def subsAll(self):
        total = 0
        current = self.head
        
        while current:
            total -= current.data
            current = current.next
        return total
    
    def addAll(self):
        total = 0
        current = self.head
        
        while current:
            total += current.data
            current = current.next
        return total
    
    def __str__(self):
        out_str = "["
        if self.head:
            current = self.head
            out_str += "%s" % current.data
            while current:
                current = current.next
                if current:
                    out_str += ", %s" % current.data
        out_str += "]"
        return out_str
    
    
```
